# Day 5 — Comprehensions
**Concepts:** list/dict/set comprehensions, conditional comprehensions

## Assignment 1
Create a list of squares of numbers 1–10 using a comprehension.

## Assignment 2
From `[1, 2, 3, 4, 5, 6, 7, 8]`, extract only odd numbers using a comprehension.

## Assignment 3
Given a list of words, create a dict of `{word: len(word)}`.

## Assignment 4
Flatten `[[1, 2], [3, 4], [5, 6]]` into a single list using a nested comprehension.

## Assignment 5
From `[1, -2, 3, -4, 5]`, create a list that replaces negatives with 0.

## Assignment 6
Create a set of unique first letters from `["apple", "avocado", "banana", "blueberry"]`.
