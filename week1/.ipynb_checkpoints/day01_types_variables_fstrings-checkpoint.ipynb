## Use of Global variable inside function:
1. use `global` to declare a golbal variable from inside a function
2. use `global` keyword to change global variable from inside a function

```python
def myFunc():
    global newG
    newG = 'new global variable from inside a func'
myFunc()
print(newG)
```

```python
# Global variables

x = 'awesome'
def myFunc():
    global x 
    x = "great"
    print('Python is ' + x)
myFunc()
print('Python is ' + x)
```

### String formatting

```python
# we cannot combine strings and numbers
age = 36
#This will produce an error:
txt = "My name is John, I am " + age
print(txt)
```

### Above oprtation can be done using f-string

```python
age = 36
#This will produce an error:
txt = f"My name is John, I am {age}"
print(txt)
```

### Whatever we place inside {} is called placeholder

We can modify the placeholder by using the : like .2f 

```python
price = 29
txt = f'The price is {price:.2f}'
print(txt)
```

```python
## you can also perform math operations inside placeholder
txt = f'mul of 2 and 3 is {2*3}'
txt
```

### Loops in python

```python
fruits = ['apple', 'banana', 'cranberry']
for f in range(len(fruits)):
    print(fruits[f])
```

```python
for fruit in fruits:
    print(fruit)
```

### List comprehension

```python
[print(x) for x in fruits]
```

```python
[print(x) for x in fruits if len(x) < 7]
```

```python
[print(x) if len(x) < 7 else print("Too long") for x in fruits]
```

### Join lists

#### Three ways:
- plus operator : To join two lists
- append() : To add single element to the list
- extend() : To add entire list to existing list

Method	  Description
append()  Adds an element at the end of the list
clear()	  Removes all the elements from the list
copy()	  Returns a copy of the list
count()	  Returns the number of elements with the specified value
extend()  Add the elements of a list (or any iterable), to the end of the current list
index()	  Returns the index of the first element with the specified value
insert()  Adds an element at the specified position
pop()	  Removes the element at the specified position
remove()  Removes the item with the specified value
reverse() Reverses the order of the list
sort()	  Sorts the list
