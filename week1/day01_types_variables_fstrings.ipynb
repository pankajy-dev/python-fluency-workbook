# Day 1 — Types, Variables, F-strings
**Concepts:** int, float, str, bool, type casting, f-strings

## Assignment 1
Create variables of each type and print them with their type using `type()`.

## Assignment 2
Take two numbers as strings `"12"` and `"5"`, cast them to int, and print their sum.

## Assignment 3
Given `name = "Alice"` and `score = 95.5`, print:
`"Alice scored 95.5 out of 100 (grade: A)"` using an f-string.

## Assignment 4
Write a one-liner that checks if `"42"` is a digit and converts it to int if so.

## Assignment 5
Print the result of `7 / 2`, `7 // 2`, and `7 % 2` — know what each returns.
