# Day 12 — Decorators
**Concepts:** functions as arguments, *args/**kwargs pass-through, @ syntax, functools.wraps

## Assignment 1
Write a `@timer` decorator that prints how long a function took to run.

## Assignment 2
Write a `@log_calls` decorator that prints the function name and arguments each call.

## Assignment 3
Write a `@retry(n)` decorator that retries a function up to n times if it raises an exception.

## Assignment 4
Apply two decorators to one function and observe the order they run.
